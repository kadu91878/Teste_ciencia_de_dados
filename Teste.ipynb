```python
#Adicionando a coluna de Mês
df['Mes'] = df['Data'].dt.to_period('M')
```

```python
#Criação de um Pandas Series com os meses e as somas dos balanços
balanco_mensal_soma = df.groupby('Mes')['Balanço Final'].sum()
balanco_mensal_soma
```

```python
#Algoritmo de Kadane
def kadane(soma_mensal):
    max_sum = soma_mensal.iloc[0]
    current_sum = soma_mensal.iloc[0]
    start = 0
    end = 0
    temp_start = 0

    for idx, i in enumerate(soma_mensal.iloc[1:], start=1):
        if i > current_sum + i:
            current_sum = i
            temp_start = idx
        else:
            current_sum += i

        if current_sum > max_sum:
            max_sum = current_sum
            start = temp_start
            end = idx

    return max_sum,soma_mensal.index[start],soma_mensal.index[end]
```

```python
#Chamada e Resultado da aplicação do Algoritmo
kadane(balanco_mensal_soma)
```

# a) Quais critérios você utilizaria para indicar se o CPF é válido em termos
estruturais? Construa um algoritmo para automatizar esse processo e realize
um teste com os seguintes CPFs, indicando aqueles que são válidos ou não:
- 217.894.500-71
- 112.622.670-05
- 695.226.930-49
- 015.897.070-51
- 311.692.760-06


```python
def validar_cpf(cpf):

    #Verificar se tem ponto e hífen
    if '.' and '-' not in cpf:
        return False

    # Remover a pontuação para fazer as verificações posteriores
    cpf = cpf.replace('.', '').replace('-', '')

    # Verificar se tem 11 dígitos
    if len(cpf) != 11:
        return False

    # Verificar se todos os dígitos são iguais
    if cpf == cpf[0] * len(cpf):
        return False

    # Calcular o primeiro dígito verificador
    soma = sum(int(cpf[i]) * (10 - i) for i in range(9))
    primeiro_digito = 11 - (soma % 11)
    primeiro_digito = 0 if primeiro_digito >= 10 else primeiro_digito

    # Calcular o segundo dígito verificador
    soma = sum(int(cpf[i]) * (11 - i) for i in range(10))
    segundo_digito = 11 - (soma % 11)
    segundo_digito = 0 if segundo_digito >= 10 else segundo_digito

    # Verificar se os dígitos são válidos
    return cpf[9] == str(primeiro_digito) and cpf[10] == str(segundo_digito)
```

```python
#Lista de exemplos dada na tarefa
cpfs = ["217.894.500-71", "112.622.670-05", "695.226.930-49", "015.897.070-51",
        "311.692.760-06"]
```

```python
#Verificação da lista de CPFs
for cpf in cpfs:
    print(f"CPF {cpf} é {'válido' if validar_cpf(cpf) else 'inválido'}")
```

# Conclusão:
Algoritmo criado baseado nas regras encontradas no seguinte site:
https://www.macoratti.net/alg_cpf.htm

# b) Quais critérios você utilizaria para indicar se um telefone celular é válido ou
não em termos estruturais? Construa um algoritmo para automatizar esse
processo e realize um teste com os seguintes telefones, indicando aqueles
que são válidos ou não:
- (64) 96743-8065
- (75) 89555-5682
- (99) 98362-0347
- (10) 98595-1389


```python
def validar_telefone(telefone):
    # Remover pontuação
    telefone = telefone.replace('(', '').replace(')', '').replace(' ', '').replace('-', '')

    ddd_nao_existente = ['23', '25', '26', '29', '36', '39', '52', '56', '57',
                         '58', '59', '72', '76', '78',]

    telefone_ddd = telefone[0] + telefone[1]


    # Verificar se tem 11 dígitos
    if len(telefone) != 11:
        return False

    #Verifica se o DDD começa com 0
    if telefone[0] == '0':
        return False

    #Verifica se o segundo digito do DDD é 0
    if telefone[1] == '0':
        return False

    #Verifica se o DDD está na lista de DDD's que não existem
    if telefone_ddd in ddd_nao_existente:
        return False


    # Verificar se o primeiro dígito depois do DDD é 9
    return telefone[2] == '9'
```

```python
telefones = ["(64) 96743-8065", "(75) 89555-5682", "(99) 98362-0347", "(10) 98595-1389" ]
```

```python
for telefone in telefones:
    print(f"Telefone {telefone} é {'válido' if validar_telefone(telefone) else 'inválido'}")
```

# Conclusão:
Algoritmo simples para realizar verificações se o número apresentado é um celular. Se começa com o número 9. Verificando também se o DDD é válido.

# c) Erros e variações de escrita são comuns em processos de preenchimento de
dados como nomes. Com isso em mente, qual estratégia você tomaria para
que em um contexto de avaliação, um modelo retornasse True para as
seguintes comparações:
- Fernando dos Santos / Fernando Santos
- Maria da Silva Rodrigues / Maria Rodrigues
- João Roberto / joao robert

```python
import unidecode

def normalize_name(name):
    # Converter para minúsculas
    name = name.lower()

    # Remover acentuação
    name = unidecode.unidecode(name)

    # Remover preposições
    prepositions = ['da', 'de', 'dos', 'do']
    name_parts = [word for word in name.split() if word not in prepositions]

    return ' '.join(name_parts)

# Lista de exemplos
names = [
    ("Fernando dos Santos", "Fernando Santos"),
    ("Maria da Silva Rodrigues", "Maria Rodrigues"),
    ("João Roberto", "joao robert")
]

```

```python
from fuzzywuzzy import fuzz

def is_similar(name1, name2, threshold=80):
    # Normalizar os nomes
    name1 = normalize_name(name1)
    name2 = normalize_name(name2)

    # Calcular similaridade fuzzy
    similarity = fuzz.ratio(name1, name2)

    return similarity >= threshold

# Testando com os exemplos
for name1, name2 in names:
    if is_similar(name1, name2):
        print(f"{name1} e {name2}: True")
    else:
        print(f"{name1} e {name2}: False")
```

# d) Analisar a estabilidade do processamento dos parceiros é um dos pontos
vitais para a manutenção da saúde processual de uma fintech, com isso em
mente, em termos percentuais e inteiros, qual foi a oscilação do total de
parceiros com processamento maior do que zero de agosto para setembro?

```python
#Filtrando o DF por tpvs positivos
df_tpv_positivo = df_integrado[df_integrado['tpv'] > 0]

#Criando o DF de tpv positivo agrupado por mês
df_tpv_positivo_por_mes = df_tpv_positivo.groupby(['mes_ano', 'user_id'])[
    'tpv'
    ].sum().reset_index()

#Contabilizando a quantidade de usuários únicos por mês com tpv positivo
df_usuarios_tpv_positivo_por_mes = df_tpv_positivo_por_mes.groupby('mes_ano')[
    'user_id'
    ].nunique().reset_index()

#Contabilizando quantidade de usuários para o mês de agosto e Setembro
usuarios_agosto = df_usuarios_tpv_positivo_por_mes[
    df_usuarios_tpv_positivo_por_mes
     [
    'mes_ano'
    ] == '2024-08']['user_id'].values[0]
usuarios_setembro = df_usuarios_tpv_positivo_por_mes[
    df_usuarios_tpv_positivo_por_mes[
    'mes_ano'
    ] == '2024-09']['user_id'].values[0]

#Calculando a porcentagem de mudança entre Agosto e Setembro
mudanca_porcentagem = ((usuarios_setembro - usuarios_agosto)
/ usuarios_agosto) *100
```

```python
(usuarios_agosto, usuarios_setembro, mudanca_porcentagem)
```
